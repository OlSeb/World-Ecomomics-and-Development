```python
def convert_million_to_numeric(value):
    if isinstance(value, str) and 'M' in value:
        numeric_value = float(value.replace('M', ''))
        return numeric_value * 1000000
    if isinstance(value, str) and 'B' in value:
        numeric_value = float(value.replace('B', ''))
        return numeric_value * 1000000000
    if isinstance(value, str) and 'k' in value:
        numeric_value = float(value.replace('k', ''))
        return numeric_value * 1000
    return value

# Apply the function to the 'Value' column
pop['2020'] = pop['2020'].apply(convert_million_to_numeric)

print(pop)
```
